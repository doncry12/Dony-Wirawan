```python
# List suhu awal dalam Celsius
temperatures = [25, 28, 30, 22, 35, 40, 18, 24]

# 1. Menggunakan list comprehension untuk konversi ke Fahrenheit
# Rumus: (C * 9/5) + 32
temp_fahrenheit = [(c * 9/5) + 32 for c in temperatures]

# 2. Mengambil 3 data terakhir menggunakan slicing
last_three_temps = temp_fahrenheit[-3:]

# Menampilkan hasil
print("List Fahrenheit:", temp_fahrenheit)
print("3 Data Terakhir:", last_three_temps)
```

```python
inventory = {
    "laptop": {"stok": 10, "harga": 15000000},
    "mouse": {"stok": 50, "harga": 25000000},
    "monitor": {"stok": 20, "harga": 3000000}
}

# 1. Tambahkan produk baru: "keyboard"
inventory["keyboard"] = {"stok": 30, "harga": 500000}

# 2. Perbarui harga "laptop" menjadi 14500000 menggunakan metode .update()
inventory["laptop"].update({"harga": 14500000})

# 3. Hitung total nilai aset di gudang (stok dikali harga untuk semua barang)
total_aset = 0
for item in inventory.values():
    total_aset += item["stok"] * item["harga"]

# Menampilkan hasil
print(f"Inventory terbaru: {inventory}")
print(f"Total nilai aset: Rp {total_aset:,}")
```

```python
import numpy as np

scores = np.array([
    [80, 85, 90],
    [70, 60, 75],
    [95, 90, 100],
    [40, 50, 45]
])

# 1. Hitung rata-rata nilai untuk setiap mahasiswa (baris)
# axis=1 berarti operasi dilakukan secara horizontal (per baris)
rerata_mahasiswa = np.mean(scores, axis=1)

# 2. Cari nilai tertinggi dari seluruh data
nilai_tertinggi = np.max(scores)

# 3. Gunakan np.where untuk status "Lulus" atau "Gagal"
# Syarat: Lulus jika rata-rata >= 70
status = np.where(rerata_mahasiswa >= 70, "Lulus", "Gagal")

# Menampilkan hasil
print("Rata-rata tiap mahasiswa:", rerata_mahasiswa)
print("Nilai tertinggi di kelas:", nilai_tertinggi)
print("Status kelulusan:", status)
```

```python
import pandas as pd
import numpy as np

df_data = {
    'Nama': ['Andi', 'Budi', 'Caca', 'Dedi', 'Euis'],
    'Departemen': ['IT', 'HR', 'IT', 'Sales', 'Sales'],
    'Gaji': [8000000, 7000000, 8500000, np.nan, 6000000],
    'Pengalaman_Tahun': [5, 3, 6, 2, 1]
}

df = pd.DataFrame(df_data)

# 1. Isi nilai Gaji yang kosong (NaN) dengan median dari seluruh gaji
median_gaji = df['Gaji'].median()
df['Gaji'] = df['Gaji'].fillna(median_gaji)

# 2. Buat kolom 'Kategori_Senioritas'
# Menggunakan np.where: jika pengalaman > 4 maka 'Senior', jika tidak 'Junior'
df['Kategori_Senioritas'] = np.where(df['Pengalaman_Tahun'] > 4, 'Senior', 'Junior')

# 3. Tampilkan rata-rata Gaji per Departemen menggunakan .groupby()
rata_gaji_dept = df.groupby('Departemen')['Gaji'].mean()

# Menampilkan hasil
print("DataFrame Terbaru:")
print(df)
print("\nRata-rata Gaji per Departemen:")
print(rata_gaji_dept)
```
